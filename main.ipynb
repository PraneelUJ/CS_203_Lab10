### Task1


```python
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest
import numpy as np
```

```python
df=pd.read_csv("ad_click_dataset.csv")
df
```

```python
df = df.dropna()
```

```python
df
```

```python
print(df['gender'].unique())
print(df['ad_position'].unique())
```

```python
df['gender'] = df['gender'].map({'Male': 0, 'Female': 1, 'Non-Binary': 2})
df['ad_position'] = df['ad_position'].map({'Top':0,'Bottom':1,'Side':2})
```

```python
df
```

```python
groupa=df[df['ad_position']==0]

groupb=df[df['ad_position']==1]
```

```python
groupa
```

```python
groupb
```

```python
a=groupa['click'].sum()
b=groupb['click'].sum()

lena=len(groupa)
lenb=len(groupb)

counting=[a,b]
lens=[lena,lenb]

zscore,pvalue=proportions_ztest(counting,lens)

print(f"Z-Score: {zscore}")
print(f"P-Value: {pvalue}")
```

### Task2


```python
from scipy.stats import ks_2samp
```

```python
df1=pd.read_csv("test1.csv")
df2=pd.read_csv("test2.csv")
df3=pd.read_csv("train.csv")
```

```python
df1
```

```python
trainNo2=df3['NO2(GT)'].dropna()
test1No2=df1['NO2(GT)'].dropna()
test2No2=df2['NO2(GT)'].dropna()
```

```python
df2
```

```python
ks1,p1=ks_2samp(trainNo2,test1No2)
ks2,p2=ks_2samp(trainNo2,test2No2)
```

```python
print(f"Test1 vs Train - KS Statistic: {ks1:.5f} --- P-Value: {p1}")
print(f"Test2 vs Train - KS Statistic: {ks2:.5f} --- P-Value: {p2}")
```
