```python
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest
import numpy as np
```

```python
df=pd.read_csv("ad_click_dataset.csv")
df
```

```python
df = df.dropna(subset=['click'])
df = df.dropna(subset=['ad_position'])
```

```python
df
```

```python
print(df['gender'].unique())
print(df['ad_position'].unique())
```

```python
# The categorical columns are gender,ad_position
df['gender'] = df['gender'].astype('category')
df['ad_position'] = df['ad_position'].astype('category')

```

```python
df['gender'] = df['gender'].map({'Male': 0, 'Female': 1, 'Non-Binary': 2})
df['gender'] = df['gender'].astype('category')
df['gender'] = df['gender'].cat.add_categories([-1])
df['gender'] = df['gender'].fillna(-1)
df['ad_position'] = df['ad_position'].map({'Top':0,'Bottom':1,'Side':2})
```

```python
df
```

```python
groupa=df[df['ad_position']==0]

groupb=df[df['ad_position']==1]
```

```python
groupa
```

```python
groupb
```

```python
a=groupa['click'].sum()

b=groupb['click'].sum()

lena=len(groupa)
lenb=len(groupb)

counting=[a,b]
lens=[lena,lenb]

zscore,pvalue=proportions_ztest(counting,lens)

print(f"Z-Score: {zscore}")
print(f"P-Value: {pvalue}")
```

Yes, there is statistically significant difference in the click rates of top and bottom ad_position

* There is a statistically significant difference in click-through rates between the two groups (p-value = 0.000048 < 0.05). We reject the null hypothesis and conclude that the difference is not due to chance.
