```python
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest
import numpy as np
```

```python
df=pd.read_csv("ad_click_dataset.csv")
df
```

```python
df = df.dropna(subset=['click'])
df = df.dropna(subset=['gender'])
df = df.dropna(subset=['ad_position'])
```

```python
df
```

```python
print(df['gender'].unique())
print(df['ad_position'].unique())
```

```python
df['gender'] = df['gender'].map({'Male': 0, 'Female': 1, 'Non-Binary': 2})
df['ad_position'] = df['ad_position'].map({'Top':0,'Bottom':1,'Side':2})
```

```python
df
```

```python
groupa=df[df['ad_position']==0]

groupb=df[df['ad_position']==1]
```

```python
groupa
```

```python
groupb
```

```python
a=groupa['click'].sum()

b=groupb['click'].sum()

lena=len(groupa)
lenb=len(groupb)

counting=[a,b]
lens=[lena,lenb]

zscore,pvalue=proportions_ztest(counting,lens)

print(f"Z-Score: {zscore}")
print(f"P-Value: {pvalue}")
```

Yes, there is statistically significant difference in the click rates of top and bottom ad_position

* As z score is -2.98 the difference between 2 groups click rates is 3 standard deviations away from the mean under null hypothesis.
* p value is 0.00285 which is much less than 0.05, hence we reject null hypothesis which stated that there is no difference in click rates in the top and bottom ad-position
